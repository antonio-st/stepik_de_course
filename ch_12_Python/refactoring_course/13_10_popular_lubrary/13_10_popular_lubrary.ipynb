`Модуль` — это файл с кодом на Python (расширение .py), который содержит функции, классы и переменные, которые можно использовать в других частях программы. Модуль представляет собой отдельный файл и позволяет организовать код в логические части, что делает его более структурированным и легко поддерживаемым.

`Библиотека` — это набор модулей, упакованных вместе и предназначенных для выполнения множества связанных задач. Библиотеки обычно содержат несколько модулей, которые вместе предоставляют функциональные возможности, например, для работы с данными, построения графиков, обработки изображений и т. д.

def mean(numbers):
    return sum(numbers) / len(numbers) if numbers else 0

def median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    mid = n // 2
    if n % 2 == 0:
        return (sorted_numbers[mid - 1] + sorted_numbers[mid]) / 2
    return sorted_numbers[mid]

    avg = mean(numbers)
    return sum((x - avg) ** 2 for x in numbers)

````python
# генерация пароля
import random
import string

def generate_password(length=8):
    characters = string.ascii_letters + string.digits + string.punctuation
    password = ''.join(random.choices(characters, k=length))
    return password

print(generate_password(12))
````

````python
import math

print(math.ceil(5.1)) # округление вверх
print(math.floor(5.9)) # округление вниз
print(math.trunc(5.8)) # отбрасывание дробного
````

`math.exp(x)` - вычисление e^x (экспонента).<br>
`math.log(x, base)` - логарифм числа x по основанию base. По умолчанию base=e.<br>
`math.log10(x)` - логарифм числа x по основанию 10.
`math.log2(x)` - логарифм числа x по основанию 2.<br>
`math.pow(x, y)` - возведение числа x в степень y.<br>
`math.sqrt(x)` - квадратный корень из x.<br>
`math.fabs(x)` - абсолютное значение числа x.<br>
`math.factorial(x)` - факториал числа x.<br>
`math.gcd(x, y)` - наибольший общий делитель чисел x и y.<br>

````python
print(math.exp(2))         # Вывод: 7.3890560989306495 (e^2)
print(math.log(8, 2))      # Вывод: 3.0 (логарифм по основанию 2)
print(math.log10(100))     # Вывод: 2.0 (десятичный логарифм)
print(math.pow(2, 3))      # Вывод: 8.0 (2^3)
````

````python
print(math.sqrt(16))       # Вывод: 4.0 (квадратный корень из 16)
print(math.fabs(-7.5))     # Вывод: 7.5 (модуль числа)
print(math.factorial(5))   # Вывод: 120 (5! = 5 * 4 * 3 * 2 * 1)
print(math.gcd(54, 24))    # Вывод: 6 (наибольший общий делитель 54 и 24)
````
